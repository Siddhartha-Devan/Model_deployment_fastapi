```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

import tensorflow as tf
```

```python

data_loader = sklearn.datasets.load_iris()

```

```python
data_loader.feature_names
```

```python
X = data_loader["data"]
Y = data_loader['target']
print(X[120], X.shape)
print(Y, Y.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

Y = encoder.fit_transform(Y.reshape((-1,1)))
```

```python
import sklearn.model_selection


X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X,Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
print(X_train.shape)
```

```python
print(Y_train, Y_test)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, Y_train)).batch(16)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, Y_test)).batch(16)
```

```python
from model_arch import IrisModel

model = IrisModel(num_classes=3)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = 'accuracy')
```

```python
for x,y in train_dataset.take(1):
    print(y)
```

```python
model.fit(train_dataset, epochs = 100)
```

```python
model.evaluate(test_dataset)
```

```python
model.save_weights("iris_classifier_1.h5")
```

```python
from sklearn.metrics import classification_report
Y_pred = model.predict(test_dataset)
```

```python
Y_pred_l = [np.argmax(i) for i in Y_pred]
```

```python
Y_test_l = [np.argmax(i) for i in Y_test]
```

```python
print(Y_pred_l)
print(Y_test_l)
```

```python
print(classification_report(Y_pred_l, Y_test_l))
```
